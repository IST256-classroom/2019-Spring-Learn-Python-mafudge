```python
text = 'carrier dome'
params = { 'q' : text, 'format' : 'json'}
url='https://nominatim.openstreetmap.org/search'
response = requests.get(url, params=params)
print(response.url)
geodata = response.json()
lat = geodata[0]['lat']
lon = geodata[0]['lon']

print(lat,lon)
```

```python
def geocode(location):
    params = { 'q' : location, 'format' : 'json'}
    url='https://nominatim.openstreetmap.org/search'
    response = requests.get(url, params=params)
    if response.ok:
        geodata = response.json()
        lat = geodata[0]['lat']
        lon = geodata[0]['lon']
        return lat, lon
    
print("Location of Syracuse, NY is ", geocode("Syracuse, NY"))
print("Location of Mount Rushmore is ", geocode("Mount Rushmore"))

```
