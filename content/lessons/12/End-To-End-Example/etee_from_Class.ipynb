```python
import pandas as pd

tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_sovereign_states_and_dependent_territories_in_Africa')
countries = tables[1]

countries.sample(1)
```

```python
# this is how you read a table in HTML when it does like the user agent
import requests

response = requests.get('https://www.cicerolittleleague.com/Default.aspx?tabid=923825')
if response.ok:
    content = response.text
    

tables = pd.read_html(content)
games = tables[1]
games.columns = [ 'game date', 'home team', 'visiting team', 'time', '', '']
games.sample()
```

```python
# read data from espn

tables = pd.read_html('http://www.espn.com/mlb/game?gameId=401074965')
stats = tables[0]
# clean up the table --
stats = stats[ ~stats['Hitters'].str.startswith("Today's at BatsFull Player ProfileNo")]

stats
```
